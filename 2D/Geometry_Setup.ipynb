```python
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
Nx, Ny = 800, 320
cx, cy = Nx//5, Ny//2
radius = Ny//7

X, Y   = np.meshgrid(np.arange(Nx), np.arange(Ny), indexing = 'ij')
R      = np.sqrt((X - cx)**2 + (Y - cy)**2)

obst   = np.zeros((Nx, Ny))
obst[R <= radius] = 1

plt.imshow(obst.T, cmap = "Greys", interpolation = 'none')
plt.xlabel('Y cell', font = 'Consolas', fontsize = 10)
plt.ylabel('X cell', font = 'Consolas', fontsize = 10)
plt.title('Obstacle map', font = 'Consolas', fontsize = 10)
plt.show()
```

```python
def save_geometry(geom, filename):
    Nx, Ny = geom.shape
    with open(filename, 'w') as file:
        file.write(f'{Nx} {Ny}\n')
        for ix in range(Nx):
            for iy in range(Ny):
                file.write(f'{ix} {iy} {int(geom[ix,iy])}\n')
    print(f'Saved to [{filename}]')
```

```python
save_geometry(obst, 'Data/test_geom_2.txt')
```
